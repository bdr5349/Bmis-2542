### <span style="background-color:#B0E0E6">Introduction</span>

The simplest model we can think of is the **linear model**, where the response or the *dependent/predicted variable*, $\textbf{y}$, depends linearly on the covariates or the *independent/predictor variables*, $\textbf{x}_i$.

In the simple linear regression, with a single variable, we describe the relationship between the predictor and the response with a straight line. The model is:

$$ \hat{\textbf{y}}  =  \beta_0 + \beta_1 \textbf{x}_1$$

* The parameter $\beta_0$ is called the constant term or the *intercept*.
* The parameter $\beta_1$ is called the *coefficient* term or slope of the regresison line.

In the case of *multiple* linear regression we extend this idea by fitting an m-dimensional hyperplane to our m predictors.

$$ \hat{\textbf{y}}  =  \beta_{0} + \beta_{1} \textbf{x}_1  + \beta_{2} \textbf{x}_2 + \beta_3 \textbf{x}_3 + \dots + \beta_m \textbf{x}_{m} $$

Despite its name, linear regression can be used to fit non-linear functions. A linear regression model is linear in the model parameters, not necessarily in the predictors. If you add non-linear transformations of your predictors to the linear regression model, the model will be non-linear in the predictors. For example, you can include quadratic and cubic transformations of predictors:

$$ \hat{\textbf{y}}  =  \beta_{0} + \beta_{1} \textbf{x}_1  + \beta_{2} {\textbf{x}_1}^2 + \beta_3 {\textbf{x}_1}^3 + \beta_{2} \textbf{x}_2+ \dots + \beta_m \textbf{x}_{m} $$


**Algorithm for estimating the parameters**:
* The most popular one is called OLS or ordinary least squares. OLS chooses parameters in a way to minimize the square of the distance between the predicted values and the actual values. It is sometimes termed as the "L2" method.
* Least absolute deviations (LAD), or minimizing the absolute value of the distance between the predicted values and actual values, is an alternative to OLS and is "robust" to the presence of outliers in data. This is sometimes called the "L1" method.
* One reason why OLS is more popular than LAD is becuase of the "easy" calculus related to the OLS approach (e.g., finding the minimum and maximum through first and second order derivates of a function). We need iterative approaches to solve LAD problems, for example, through linear programming. Also, LAD does not offer convenient procedures for examining variances and hypothesis testing.

### <span style="background-color:#B0E0E6">Regression Analysis with Statsmodels</span>

```python
y=advertising['Sales']
X=advertising[['TV', 'Radio', 'Newspaper']]
X=sm.add_constant(X)
model1 = sm.OLS(y, X) 
results1 = model1.fit()
results1.summary()
```

<div style="text-align: right"> <a href='#section_id5'>Compare with Scikit model results</a></div>

#### Comparing the predicted values with actual values

```python
predictions1=results1.predict(X)
predictions1.head()
```

```python
advertising['predicted_sales']=predictions1
```

```python
g=sns.relplot(x="Sales", y="predicted_sales", data=advertising) #store the plot in "g"
g.set(xlabel="Actual Sales", ylabel="Predicted Sales") #set lables for axes
mn = min(advertising.Sales.min(), advertising.predicted_sales.min()) #generate min coordinates for line
mx = max(advertising.Sales.max(), advertising.predicted_sales.max()) #generate max coordinates for line
points = np.linspace(mn, mx, 100) #generate points of the line
plt.plot(points, points, color='r', linestyle='--', linewidth=1.5) #add the line to the plot
plt.show() #show the full plot
#plt.gca().plot(points, points, color='r', linestyle='--', linewidth=1.5) #also works
```

#### Evaluate prediction performance using RMSE

```python
sm.tools.eval_measures.rmse(advertising.Sales, advertising.predicted_sales, axis=0)
```

```python
np.sqrt(np.square(advertising.Sales-advertising.predicted_sales).mean()) 
```

<Mark>Interpretation:</Mark> Suppose, sales was measured in millions of dollars, our prediction is deviant from the actual values by about $ 1.7 million, on average.<br> 
Whether that accuracy of prediction is good or not is a subjective decision, depending on the business context.

```python
# We can also use this handy function from the DMBA package to print all the metrics in one go.
regressionSummary(advertising.Sales, advertising.predicted_sales)
```

#### Regression diagnostics

```python
residuals=results1.resid #provided by statsmodels
# residuals=advertising.Sales-advertising.predicted_sales #manual calculation
```

#### Normality of residuals

```python
sns.distplot(residuals, bins=10, kde=True, hist=True, fit=stats.gamma, fit_kws={"color":"red"})
```

* The assumption related to the normality of the residuals does seem to be violated.
* The q-q plot is another way to visualize this. If the residuals are normally distributed, they should all line up along the red reference line in the chart below.

```python
fig=sm.qqplot(residuals, stats.t, fit=True, line='45')
plt.show()
```

More formal tests of the normality assumption:
* Jarque-Bera test compares the skewness and kurtosis of the residuals with the normal distribution. The null hypothesis is that the residuals are normally distributed. So a desirable result is to get a p-value of >0.05 in the hypothesis test using the JB test.
* Skewness of the normal distribution is 0.
* Kurtosis of the normal distribution is 3. 
* The omni_normtest is similar to the JB test and uses both kurtosis and skewness to formulate a test statistic. Once again, the null hypothesis is that the residuals are normally distributed.

```python
name = ['Jarque-Bera', 'Chi^2 two-tail prob.', 'Skew', 'Kurtosis']
test = sms.jarque_bera(results1.resid)
lzip(name, test)
```

```python
name = ['Chi^2', 'Two-tail probability']
test = sms.omni_normtest(results1.resid)
lzip(name, test)
```

#### Heteroskedasticity

* Null hypothesis in these tests is homoskedasticity or the residual variance does not depend on the variables in the X matrix.
* In the BP test, heteroskedasticity is treated as a linear function of all the independent variables in the model.
* In the GQ test, sample observations are divided into two groups, and evidence of heteroskedasticity is based on a comparison of the residual sum of squares (RSS) using the F-statistic.

```python
name = ['Lagrange multiplier statistic', 'p-value', 'f-value', 'f p-value']
test = sms.het_breuschpagan(results1.resid, results1.model.exog)
lzip(name, test)
```

```python
name = ['F statistic', 'p-value']
test = sms.het_goldfeldquandt(results1.resid, results1.model.exog)
lzip(name, test)
```

#### Multicollinearity
* Condition number is a metric that tries to capture the variables that are similarly explaining high proportions of variances in the response variable.
* Condition numbers above 30 usually indicate the presence of multicollinearity.

```python
np.linalg.cond(results1.model.exog)
```

#### Outliers
* Influential observations can be identified through the leverage plot.
* As a sensitivity test, we can try to omit these influential observations and see if the hypothesis testing results vary.

```python
test_class = OLSInfluence(results1)
test_class.dfbetas[:5,:]
```

```python
# avoid plotting this graph if you have a large dataset!
fig, ax = plt.subplots(figsize=(8,6))
fig = plot_leverage_resid2(results1, ax = ax)
```

### <span style="background-color:#B0E0E6">K-fold Cross Validation</span>

* To simulate varying the data, we split it into training and test portions
* We are using 75% of data for "training" and 25% for validation/testing
* Random state parameter helps us to achieve the same "random" split across different computers. This can be any number. As long as we use the same number across different computers, we should get the same results.

```python
X_train, X_test, Y_train, Y_test = train_test_split(X3, Y, test_size=0.25, random_state=23)
print(X_train.shape, Y_train.shape)
print(X_test.shape, Y_test.shape)
```

```python
#training model
salesreg2=sklm.LinearRegression()
salesreg2.fit(X_train, Y_train)
```

```python
#3-fold cross validation
cvparam = KFold(3, random_state=13)
scores_mse_lm =  cross_validate(salesreg2, X_train, Y_train, cv=cvparam, scoring='neg_mean_squared_error')
```

<Mark>Note:</Mark> 
* The features of the [cross_validate() function](https://scikit-learn.org/stable/modules/cross_validation.html#cross-validation-and-model-selection) are quite extensive. Several ways of splitting the data are possible. Here, we are just doing a basic K-fold split (k=3 in our example).
* The MSE values from each of the three folds are stored in the scores_mse_lm array.
* The negative sign is becuase the parameter used by the algorithm is "neg_mean_squared_error", to indicate that a lower value of MSE is better. I multipled it by -1 to get the actual MSE.

```python
scores_mse_lm
```

```python
scores_mse_lm['test_score']
```

```python
np.sqrt(scores_mse_lm['test_score'].mean()*-1)
```

<Mark>Interpretation:</Mark> Once we induce some variation in the training sample, by doing a 3-fold split, the accuracy of our prediction has gone down. This is an example of how prediction models suffer from the "variance" problem. <br> 
Here, the 3-fold cross-validation reveals that the model is deviant from reality by about $1.73M dollars.

#### Checking prediction on the out-of-training-sample (i.e., the test sample):

```python
lm_predictions_tt=salesreg2.predict(X_test) #use test sample and coefficients derived from trained model
np.sqrt(np.square(lm_predictions_tt-Y_test).mean()) #compare predictions with actual values in test sample
```

<Mark>Interpretation:</Mark> Our model is deviant from the actual values by about 1.6 million $, on average, when using the test data. 
Is that good enough? That is indeed a subjective decision, depending on the business context.

### <span style="background-color:#B0E0E6">Non-parametric approach: KNN</span>
* We'll use the [Nearest Neighbors](https://scikit-learn.org/stable/modules/neighbors.html) module of scikit-learn
* [KNeighborsRegressor](https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsRegressor.html#sklearn.neighbors.KNeighborsRegressor) assigns values based on the mean of the values of the nearest neighbors to a data point.

```python
advertising = pd.read_csv('advertising.csv')
```

```python
# First, scale the predictors:
Xp=preprocessing.scale(advertising[['TV', 'Radio', 'Newspaper']]).astype('float64')
Yp=preprocessing.scale(advertising.Sales).astype('float64')
# not scaling the dependent variable will help comparison with the earlier models
Ypns=(advertising.Sales).astype('float64')
```

```python
# Set aside a portion (e.g., 25%) of data for out-of-training-sample testing.

# to use the scaled Sales DV:
X3p_train, X3p_test, Y3p_train, Y3p_test = train_test_split(Xp, Yp, \
                                                            test_size=0.25, random_state=7)

#using the unscaled Sales DV
X3p_train, X3p_test, Y3pns_train, Y3pns_test = train_test_split(Xp, Ypns, \
                                                            test_size=0.25, random_state=7)
print("Scaled version:")
print(X3p_train.shape, Y3p_train.shape)
print(X3p_test.shape, Y3p_test.shape)
print("Not scaled version:")
print(X3p_train.shape, Y3pns_train.shape)
print(X3p_test.shape, Y3pns_test.shape)
```

```python
# Define the KNN model parameters
# use 4 nearest neighbors weighted uniformly
knn_rcv = KNeighborsRegressor(n_neighbors=4, weights='uniform')
knn_rcv_ns = KNeighborsRegressor(n_neighbors=4, weights='uniform') 
```

More parameters to tweak are available in [the module](https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsRegressor.html), and we can use it to fine tune the models. We'll cover module tuning in Session-12.

```python
# fit the KNN model using training sample
knn_rcv.fit(X3p_train, Y3p_train)
knn_rcv_ns.fit(X3p_train, Y3pns_train)
```

```python
# how does the model perform in training?
#get the predictions using training data
knn_prediction_reg=knn_rcv.predict(X3p_train)
knn_prediction_reg_ns=knn_rcv_ns.predict(X3p_train) 
```

```python
#average error of prediction in training data: scaled version
np.sqrt(np.square(Y3p_train-knn_prediction_reg).mean()) 
```

```python
#average error of prediction in training data: not scaled version.
np.sqrt(np.square(Y3pns_train-knn_prediction_reg_ns).mean())
```

```python
import statsmodels.api as sm
sm.tools.eval_measures.rmse(Y3pns_train, knn_prediction_reg_ns, axis=0)
```

* I'll use the ToyotaCorolla dataset that is featured in the [Chapter 6 of the DMBA book](https://pitt.primo.exlibrisgroup.com/permalink/01PITT_INST/e8h8hp/alma9996364823406236)

```python
car_df = pd.read_csv('ToyotaCorolla.csv')
```

```python
car_df.info()
```

```python
#setting response and predictor variables, matching with the book chapter
predictors = ['Age_08_04', 'KM', 'Fuel_Type', 'HP', 'Met_Color', 'Automatic', 'CC', 
              'Doors', 'Quarterly_Tax', 'Weight']
outcome = 'Price'
```

```python
car_df[predictors].head()
```

Fuel_Type is a categorical predictor and needs the creation of "dummy" variables. We can do that by using the [pd.get_dummies() pandas function](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.get_dummies.html)

```python
# partition data
X = pd.get_dummies(car_df[predictors], drop_first=True)
y = car_df[outcome]
train_X, valid_X, train_y, valid_y = train_test_split(X, y, test_size=0.4, random_state=1)
```

```python
X
```

#### Exhaustive search
* Metrics such as adjusted R-squared, Akaike Information Criteria (AIC), or Bayesian Information Criteria (BIC) can be used to pick the best model from the oens that use a different subsets of predictor variables.
* We'll use a function from the DMBA package for implementing exhaustive search, and use AIC in the following example.

```python
def train_model(variables):
    model = sklm.LinearRegression()
    model.fit(train_X[variables], train_y)
    return model

def score_model(model, variables):
    pred_y = model.predict(train_X[variables])
    # we negate as score is optimized to be as low as possible
    return -adjusted_r2_score(train_y, pred_y, model)
```

```python
allVariables = train_X.columns
results = exhaustive_search(allVariables, train_model, score_model)
```

```python
data = []
for result in results:
    model = result['model']
    variables = result['variables']
    AIC = AIC_score(train_y, model.predict(train_X[variables]), model)
    
    d = {'n': result['n'], 'r2adj': -result['score'], 'AIC': AIC}
    d.update({var: var in result['variables'] for var in allVariables})
    data.append(d)
```

```python
searchresults=pd.DataFrame(data, columns=('n', 'r2adj', 'AIC') + tuple(sorted(allVariables)))
```

```python
searchresults
```

```python
sns.relplot(x="n", y="r2adj", data=searchresults, kind="line")
```

* The model with 8 predictors is preferred. 
* Age_08_04 is a dominant predictor and featured in all models

#### Backward elimination of variables
* We start with the "full" model and eliminate in steps the least useful predictor. 
* The elimination stops when all the remaining predictors make significant contributions to the model.

```python
def train_model(variables):
    model = sklm.LinearRegression()
    model.fit(train_X[variables], train_y)
    return model

def score_model(model, variables):
    return AIC_score(train_y, model.predict(train_X[variables]), model)

best_model, best_variables = backward_elimination(train_X.columns, train_model, score_model, verbose=True)

print(best_variables)
```

```python
regressionSummary(valid_y, best_model.predict(valid_X[best_variables]))
```

<div style="text-align: right"> <a href='#section_id0'>Back to top of Notebook</a></div>

#### Forward selection
* We start with the "null" model and keep adding predictors one-by-one. 
* Each added predictor will be the one that makes the best contribution to AIC.
* The algorithm stops when additional predictors don't make significant contribution.

```python
# The initial model is the constant model - this requires special handling
# in train_model and score_model
def train_model(variables):
    if len(variables) == 0:
        return None
    model = sklm.LinearRegression()
    model.fit(train_X[variables], train_y)
    return model

def score_model(model, variables):
    if len(variables) == 0:
        return AIC_score(train_y, [train_y.mean()] * len(train_y), model, df=1)
    return AIC_score(train_y, model.predict(train_X[variables]), model)

best_model, best_variables = forward_selection(train_X.columns, train_model, score_model, verbose=True)

print(best_variables)
```

<div style="text-align: right"> <a href='#section_id0'>Back to top of Notebook</a></div>

#### Stepwise selection
* It works like forward selection, but we drop variables that are not useful (similar to backward elimination).

```python
best_model, best_variables = stepwise_selection(train_X.columns, train_model, score_model, verbose=True)
print(best_variables)
```

### <span style="background-color:#B0E0E6">Lasso and Ridge Regression Approaches</span>
* These methods constrain the combined magnitude of the coefficients and "shrink" some coefficients towards zero. Coefficients of correlated predictors tend to produce high standard erros (or instability), which reduces predictive power. Small changes in data tend to radically shift which of these predictors are emphasized in the model.
* Lasso uses L1 penalty or the sum of absolute values of the coefficients
* Ridge uses L2 penalty or the sum of the the squared values of the coefficients
* We can let the algorithm optimally set the penalty, or alpha, through cross-validation

```python
lasso_cv = sklm.LassoCV(normalize=True, cv=5)
lasso_cv.fit(train_X, train_y)
regressionSummary(valid_y, lasso_cv.predict(valid_X))
print('Lasso-CV chosen regularization: ', lasso_cv.alpha_)
print(lasso_cv.coef_)
```

```python
ridge_cv = sklm.RidgeCV(normalize=True, cv=5)
ridge_cv.fit(train_X, train_y)
regressionSummary(valid_y, ridge_cv.predict(valid_X))
print('Ridge-CV chosen regularization: ', ridge_cv.alpha_)
print(ridge_cv.coef_)
```

```python
# for comparison, we can gather the estimates from OLS:
linearRegression = sklm.LinearRegression(normalize=True).fit(train_X, train_y)
regressionSummary(valid_y, linearRegression.predict(valid_X))
print(linearRegression.coef_)
```

```python
pd.DataFrame({'features': train_X.columns, 'linear regression': linearRegression.coef_, 
              'LassoCV': lasso_cv.coef_, 'RidgeCV': ridge_cv.coef_})
```

* The auto selection of the penalty in Lasso and Ridge did not eliminate any variables. 
* Lasso and Ridge did improve the prediction accuracy (lower RMSE)
